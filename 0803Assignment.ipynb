The Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a specific value. It provides a way to model and visualize the distribution of continuous data.

Q2. What are the types of Probability distribution?

Ans:

Discrete Probability Distributions:

Definition: Describe the likelihood of discrete outcomes (countable outcomes).
Characteristics: Each possible outcome has a specific probability; the sum of all probabilities equals 1.
Common Examples:
Binomial Distribution: Number of successes in a fixed number of Bernoulli trials. Example: Number of heads in 10 coin tosses.
Poisson Distribution: Number of events in a fixed interval of time or space. Example: Number of emails received in an hour.
Geometric Distribution: Number of trials until the first success. Example: Number of dice rolls until a six appears.

Continuous Probability Distributions:

Definition: Describe the likelihood of continuous outcomes (any value within a range).
Characteristics: Probabilities are described by a probability density function (PDF); area under the PDF curve represents the probability of outcomes within that range.
Common Examples:
Normal Distribution: Data clusters around a mean with a bell-shaped curve. Example: Heights of people.
Exponential Distribution: Time between events in a Poisson process. Example: Time until a light bulb burns out.
Uniform Distribution: All outcomes are equally likely within a range. Example: Rolling a fair die.
Beta Distribution: Distribution of probabilities for a variable between 0 and 1. Example: Probability of success in a Bernoulli trial.

Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import numpy as np

def normal_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = np.exp(- (x - mean)**2 / (2 * std_dev**2))
    pdf_value = coefficient * exponent
    return pdf_value

mean = 0
std_dev = 1
x = 1
pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value at x = {x} is {pdf_value:.4f}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Ans:

Properties of Binomial Distribution
Fixed Number of Trials:

The experiment is conducted a fixed number of times (n).
Two Possible Outcomes:

Each trial results in one of two outcomes: success or failure.
Constant Probability of Success:

The probability of success (p) remains constant across trials.
Independence of Trials:

The outcome of one trial does not affect the outcome of another trial.
Probability of Exactly k Successes:

The distribution describes the probability of achieving exactly k successes out of n trials.


Examples of Binomial Distribution
Coin Tosses:

Example: Calculating the probability of getting exactly 3 heads in 5 tosses of a fair coin. Here, n = 5 (number of tosses), p = 0.5 (probability of heads), and k = 3 (number of heads).
Quality Control:

Example: Determining the probability of finding exactly 2 defective items in a sample of 10 items, where each item has a 0.1 probability of being defective. Here, n = 10 (number of items), p = 0.1 (probability of a defective item), and k = 2 (number of defective items).

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10  # Number of trials
p = 0.4  # Probability of success
size = 1000  # Sample size

sample = np.random.binomial(n, p, size)

plt.figure(figsize=(10, 6))
plt.hist(sample, bins=range(n+2), edgecolor='black', alpha=0.75)
plt.title('Histogram of Binomial Distribution Sample')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(mean, x):
    return poisson.cdf(x, mu=mean)

mean = 3.5
x = 4
cdf_value = poisson_cdf(mean, x)
print(f"The CDF value at x = {x} is {cdf_value:.4f}")
```

Q7. How Binomial distribution different from Poisson distribution?

Ans:

Binomial Distribution

Nature of Distribution:

Discrete probability distribution.
Models the number of successes in a fixed number of trials, each with the same probability of success.

Parameters:

Number of trials (n).
Probability of success (p).

Number of Trials:

Fixed number of trials (n) is known.
Example: Number of heads in 10 coin tosses.

Probability of Success:

Each trial has a fixed probability (p) of success.
Example: Probability of rolling a 6 on a die in each roll.

Distribution Shape:

Can vary; can be symmetric or skewed depending on n and p.
Example: Symmetric with n = 10 and p = 0.5.

Applicability:

Suitable for scenarios with a fixed number of trials where each trial is independent.
Example: Tossing a coin 20 times and counting the number of heads.

Poisson Distribution

Nature of Distribution:

Discrete probability distribution.
Models the number of events occurring in a fixed interval of time or space, given a constant mean rate.

Parameters:

Mean rate of occurrence (lambda).

Number of Trials:

No fixed number of trials; models events in continuous intervals.
Example: Number of phone calls received by a call center in an hour.

Probability of Success:

Uses average rate (lambda) of events, not a fixed probability.
Example: Average number of emails received per day.

Distribution Shape:

Often right-skewed, especially for smaller lambda; approaches normal distribution as lambda increases.
Example: Right-skewed with lambda = 2.

Applicability:

Suitable for modeling the number of events in continuous intervals of time or space.
Example: Modeling the number of cars passing through a toll booth in an hour.

Relation to Binomial:

Poisson distribution can approximate the Binomial distribution when the number of trials (n) is large and the probability of success (p) is small, with np being moderate.
Example: Counting rare defects in a large batch.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

mean = 5
sample_size = 1000

sample = np.random.poisson(lam=mean, size=sample_size)

sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

Mean and Variance in Binomial Distribution

Mean:

Formula: Mean = Number of trials × Probability of success
Description: Represents the average number of successes in a fixed number of trials.

Variance:

Formula: Variance = Number of trials × Probability of success × (1 - Probability of success)
Description: Measures how much the number of successes varies around the mean.


Mean and Variance in Poisson Distribution

Mean:

Formula: Mean = Average rate of occurrences
Description: Represents the expected number of events in a given interval.

Variance:

Formula: Variance = Average rate of occurrences
Description: Shows how much the number of occurrences varies around the average rate.
