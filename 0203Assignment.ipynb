Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Ans:
A scatter plot in matplotlib is a 2d-plot  that represts the scatterign of the data on cartesian plane.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

#taking common x,
x=x1

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y1, 'r-o')  # 'r-o' means red color with circle markers
plt.title('Line 1')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g-s')  # 'g-s' means green color with square markers
plt.title('Line 2')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b-^')  # 'b-^' means blue color with triangle markers
plt.title('Line 3')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'm-')  # 'm-' means magenta color with line only
plt.title('Line 4')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Ans:
Bar plot represents the data indivisually as category or groups.
It helps in better visualization, better analysis of categorical data

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(company, profit, color='skyblue')
plt.title('Vertical Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.xticks(rotation=45)  

plt.subplot(1, 2, 2)
plt.barh(company, profit, color='salmon')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Ans:
Box plot represents the distrbution of data through quartiles. It is popular because of its ability to visually summarising the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.ylabel('Value')

plt.show()
```
