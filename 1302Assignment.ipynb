Q5. Explain ImportError. What is ModuleNotFoundError?

A:
ImportError occurs when an imported class is unavailable or not created or misspelled.
ModuleNotFoundError occurs when the module doesn't exist or misspelled or path is incorrect.

Q6. List down some best practices for exception handling in python.

A:
1. Use Exception for exceptional cases
1. Don't swallow the exception
1. catch specific exceptions
1. cleanup the resources in 'finally' block
1. raise custom exception
1. document all exceptions thrown by a function
