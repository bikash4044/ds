Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:



scatter plot: Displays the relationship between two continuous variables, useful for identifying correlations

Histogram: Shows the distribution of a single continuous variable by dividing the data into bins and plotting frequencies

Box plot: Summarizes the distribution of a continuous variable through its quartiles and median, highlighting potential outliers

Pair Plot: Provides a matrix of scatter plots and histograms for pairwise relationships and distributions of multiple numerical variables

Heatmap: Visualizes data in a matrix format with color-coding to show the magnitude of values, useful for understanding correlations

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("fmri")

sns.lineplot(data=data, x="timepoint", y="signal", hue="event", style="region")

plt.title('Line Plot of Signal over Timepoint by Event and Region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("titanic")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  
sns.boxplot(data=data, x="pclass", y="age")
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1, 2, 2) 
sns.boxplot(data=data, x="pclass", y="fare")
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")

sns.histplot(data=data, x="price", hue="cut", bins=30, kde=False)

plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")

sns.pairplot(data=data, hue="species")
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)

plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")

pivot_data = data.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot_data, annot=True, cmap="jet", fmt="d")
plt.title('Heatmap of Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
