### Q1. Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.

```python
import pandas as pd
s= pd.Series([4,8,15,16,23,42])
print(s)
```

### Q2. Create a variable of list type containing 10 elements in it, and apply pandas.Series function on the variable print it.

```python
import pandas as pd
s1 = [i+1 for i in range(10)]
s = pd.Series(s1)
print(s)
```

### Q3. Create a Pandas DataFrame that contains the following data:

| Name   | Age  | Gender |
| -----  | ---- | ------ |
| Alice  | 25   | Female |
| Bob    | 30   |  Male  |
| Claire | 27   | Female |

### Then, print the DataFrame.

```python
import pandas as pd
name = ["Alice","Bob","Claire"]
age = [25,30,27]
gender = ["Female","Male","Female"]
df = pd.DataFrame()
df["NAME"] = name
df["AGE"] = age
df["GENDER"] = gender
print(df)
```

### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A:
Dataframe is a two dimensional data structure in pandas module. pandas.series is a one dimensional data structure.
a data structure created using pd.DataFrame() such as in question 3, is a dataframe. A series is created using pd.Series(), as shown in question 1 and 2.

```python
import pandas as pd
name = ["Alice","Bob","Claire"]
age = [25,30,27]
gender = ["Female","Male","Female"]
df = pd.DataFrame()
df["NAME"] = name
df["AGE"] = age
df["GENDER"] = gender
print("Dataframe:\n",df)
s = pd.Series(name)
print("\nSeries:\n",s)
```

### Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

A:
    
### 1. Series
- import pandas as pd

- Create a Series
- my_series = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])

- Modify an element
- my_series['b'] = 10

- Add a new element
- my_series['e'] = 5

- Delete an element
- del my_series['c']


### 2. Dataframe

- import pandas as pd

- Create a DataFrame
- df = pd.DataFrame({'A': [1, 2, 3], 'B': ['a', 'b', 'c']})

- Modify a value
- df.at[1, 'A'] = 20

- Add a new column
- df['C'] = [4, 5, 6]

- Remove a column
- del df['B']

- Add a new row
- new_row = pd.Series({'A': 4, 'C': 7})
- df = df.append(new_row, ignore_index=True)


### 3. Panel:
 considering Panel in python, it is just a multiple layers od dataframe, since dataframe is mutable, we also can alter the content of a panel.
 
 
### Hence all of Series, Dataframe and Panel are MUTABLE.
